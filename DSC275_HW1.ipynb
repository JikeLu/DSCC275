```python
plt.figure(figsize=(16, 8), dpi=200)
plt.plot(Air_df["Vehicle"])
plt.show()
```

```python
from statsmodels.tsa.seasonal import seasonal_decompose
```

```python
result_air=seasonal_decompose(Air, model='multiplicable', period=12)
result_air.plot();
```

## (b)

```python
plt.figure(figsize=(16, 8), dpi=200)
plt.plot(Air_df["Rail"])
plt.show()
```

```python
result_rail=seasonal_decompose(Air_df["Rail"], model='multiplicable', period=12)
result_rail.trend.plot()
```

The trend of Rail time series have changed over time. Initially(first 25%), the mean is constant; then(25% to 50%),the trend tends to decline; finally(50% to 100%), the trend has increased. Overall, the trend is decreasing, which means the mean at the end is lower than initial.

## (a)

```python
Coal_df = pd.read_csv("Coal_Production_US_2020.csv")
```

```python
plt.figure(figsize=(16, 8), dpi=200)
Coal_df['Date'] = pd.to_datetime(Coal_df['Year'],format="%Y")
Coal_df = Coal_df.drop(columns = 'Year')
Coal_df = Coal_df.set_index('Date')
plt.plot(Coal_df['Coal Production, Short Tons in Thousands'], color = 'cornflowerblue')
plt.xlabel('Year', size = 12)
plt.ylabel('Coal Production, Short Tons in Thousands', size = 12)
plt.show()
```

```python
plot_acf(Coal_df['Coal Production, Short Tons in Thousands'], lags=30);
```

## (b)

From the time series graph, we can see it is non-stationary, since it has a trend and has no constant mean.

## (c)

```python
first_diff=[]
for i in range(len(Coal_df)-1):
    first_diff.append(Coal_df['Coal Production, Short Tons in Thousands'][i+1]-Coal_df['Coal Production, Short Tons in Thousands'][i])
```

```python
plt.figure(figsize=(16, 8), dpi=200)
plt.plot(first_diff)
plt.show()
```

```python
plot_acf(first_diff, lags=30);
```

## (d)

one important impact of differencing on time series is that it can help us remove trend, which means it can make a non-stationary time series become stationary.

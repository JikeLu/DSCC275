# Q4

```python
BS = pd.read_csv('Beverages_Shipment_2020.csv')
BS['Month'] = pd.to_datetime(BS['Month'], format="%b-%y")
```

```python
plt.plot(BS['Month'],BS["Dollars (in Millions)"], '.-')
plt.xlabel('Time')
plt.ylabel('Dollars in Millions')
plt.title('Beverages Shipments Time Series')
plt.show()
```

*(a)* There is a seasonality in this pattern.

*(b)* Yes, there is a seasonality in this time series.

```python
from statsmodels.graphics.tsaplots import plot_acf
plot_acf(BS["Dollars (in Millions)"], lags = 30)
plt.title('Beverages Shipments ACF')
plt.show()
```

*(c)* The seasonal period is about 1 year because a peak appears every 12 lags.

# Q5

```python
CP = pd.read_csv('Coal_Production_US_2020.csv')
CP['Year'] = pd.to_datetime(CP['Year'], format="%Y")
```

```python
plt.plot(CP['Year'],CP["Coal Production, Short Tons in Thousands"], '.-')
plt.xlabel('Time')
plt.ylabel('Coal Production, Short Tons in Thousands')
plt.title('Coal Production Time Series')
plt.show()
```

```python
plot_acf(CP["Coal Production, Short Tons in Thousands"], lags = 56)
plt.title('Coal Production ACF')
plt.show()
```

*(b)* It is non-stationary because it clearly shows an increasing trend.

```python
cp_list = CP['Coal Production, Short Tons in Thousands'].tolist()
time_list = CP['Year'].tolist()
time_list = time_list[1:]
first_diff = list()
for i in range(0,len(cp_list)-1):
    num = cp_list[i+1] - cp_list[i]
    first_diff.append(num)
```

```python
plt.plot(time_list,first_diff, '.-')
plt.title('First Difference')
```

```python
plot_acf(first_diff, lags = 55)
plt.title('First Difference ACF')
plt.show()
```

*(d)* Differencing the time series makes it appear to be stationary. The first difference also tells how the data flunctuates between time periods.
